```python
def daily_PM25_data_extraction(data):
    data = np.array(data)   
    daily_data = []
    for i in range(8760 // 24):
        nw = np.mean(data[:, i * 24 : (i + 1) * 24], axis=1)
        daily_data.append(nw)
    daily_PM_data = pd.DataFrame(daily_data)
    return daily_PM_data.T
```
